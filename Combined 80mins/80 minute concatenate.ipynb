```python
import os
import pandas as pd
import numpy as np
```

```python
# Define the base paths and file names
base_paths = [
    '/Users/priyank/Desktop/ML4QS-project/All DATA/LOF Processed Data/',
    '/Users/priyank/Desktop/ML4QS-project/All DATA V2/LOF Processed Data/',
    '/Users/priyank/Desktop/ML4QS-project/Caleb_raw_data/',
    '/Users/priyank/Desktop/ML4QS-project/Rex_raw_data/'
]
file_names_v1 = [
    'Combined_Accelerometer_0.5s_interpolated_spline.csv',
    'Combined_Gyroscope_0.5s_interpolated_spline.csv',
    'Combined_Linear Accelerometer_0.5s_interpolated_spline.csv',
    'Combined_Magnetometer_0.5s_interpolated_spline.csv'
]
file_names_v2 = [
    'Combined_accelerometer_interpolated_spline.csv',
    'Combined_gyroscope_interpolated_spline.csv',
    'Combined_linear_accelerometer_interpolated_spline.csv',
    'Combined_magnetometer_interpolated_spline.csv'
]
file_names_v3 = [
    'Combined_accelerometer_20min.csv',
    'Combined_gyroscope_20min.csv',
    'Combined_linear_accelerometer_20min.csv',
    'Combined_magnetometer_20min.csv'
]
file_names_v4 = [
    'Combined_magnetometer_20min.csv',
    'Combined_linear_accelerometer_20min.csv',
    'Combined_gyroscope_20min.csv',
    'Combined_accelerometer_20min.csv'
]
sensor_types = ['accelerometer', 'gyroscope', 'linear_accelerometer', 'magnetometer']
gesture_labels = ['handshake', 'highfive', 'clapping', 'waving']  # Corresponding labels for gestures

# Function to load data from both paths and combine them
def load_and_combine_data(sensor_type, file_names_v1, file_names_v2, file_names_v3, file_names_v4):
    combined_data = pd.DataFrame()
    time_offset = 0
    for label in gesture_labels:
        for base_path, file_names in zip(base_paths, [file_names_v1, file_names_v2, file_names_v3, file_names_v4]):
            file_name = file_names[sensor_types.index(sensor_type)]
            file_path = os.path.join(base_path, file_name)
            data = pd.read_csv(file_path)

            # Filter data for the specific label
            start_idx = gesture_labels.index(label) * 300 * 2
            end_idx = (gesture_labels.index(label) + 1) * 300 * 2
            gesture_data = data.iloc[start_idx:end_idx].copy()

            # Correct the time column
            gesture_data['time'] = np.round(np.arange(time_offset, time_offset + len(gesture_data) * 0.5, 0.5), 1)
            gesture_data['label'] = label

            combined_data = pd.concat([combined_data, gesture_data])
            time_offset += len(gesture_data) * 0.5

    return combined_data.reset_index(drop=True)

# Combine data for each sensor type
combined_dfs = {}
for sensor_type in sensor_types:
    combined_dfs[sensor_type] = load_and_combine_data(sensor_type, file_names_v1, file_names_v2, file_names_v3, file_names_v4)

# Save the combined data
for sensor_type, combined_df in combined_dfs.items():
    output_file = f'Combined_{sensor_type}_80min.csv'
    combined_df.to_csv(output_file, index=False)
    print(f"Saved combined data for {sensor_type} to {output_file}")

print("Data combination and label assignment completed.")
```
